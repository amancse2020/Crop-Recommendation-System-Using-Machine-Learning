```python
import pandas as pd
import numpy as np
```

```python
from scipy.sparse.construct import random
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Reading the csv file
data=pd.read_csv('crops_final.csv')
#data=pd.read_csv('../dataset/temp_humid_crop.csv')
# print(data.head(1))
# print(data.crop)
```

```python
# To add state uncomment the next line
#state= pd.get_dummies(data.State_Name).iloc[: , :]
# season = pd.get_dummies(data.Season).iloc[: , :]
# Creating dummy variable for target i.e label
label= pd.get_dummies(data.crop).iloc[: , :]
#season= pd.get_dummies(data.Season).iloc[: , :]
# print(season)
#data= pd.concat([data,state,label],axis=1)
data= pd.concat([data,label],axis=1)
#print(data)
data.drop('crop', axis=1,inplace=True)
# data.drop('Sowing Temp', axis=1,inplace=True)
# data.drop('Harvest Temp', axis=1,inplace=True)
#print(data)
print('The data present in one row of the dataset is')
print(data.head(1))
train=data.iloc[:, 6:14].values  # 41, if states are included
test=data.iloc[: ,14:].values
print(train[0],test[0])
```

```python
#Dividing the data into training and test set
X_train,X_test,y_train,y_test=train_test_split(train,test,test_size=0.2,random_state=195)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Importing Decision Tree classifier
from sklearn.tree import DecisionTreeRegressor
clf=DecisionTreeRegressor()
```

```python
# Importing Random Forest classifier
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators=400, min_samples_split=100,oob_score=True, verbose=4, n_jobs=4, random_state=7, warm_start=True, max_features=4)
```

```python
#Fitting the classifier into training set
clf.fit(X_train,y_train)
pred=clf.predict(X_test)
```

```python
test_id_no = 2000
print(X_test[test_id_no], pred[test_id_no] , y_test[test_id_no])

print("predicted crop:- ", np.argmax(pred[test_id_no]))
print("Actual crop:- ", np.argmax(y_test[test_id_no]))
```

```python
# Testing on custom input
crops = ['arhar', 'bajra', 'barley', 'coriander', 'cotton', 'cowpea', 'dry chillies', 'garlic', 'ginger', 'gram', 'groundnut', 'jowar', 'linseed', 'maize-k', 'maize-r', 'masoor', 'moong', 'onion', 'peas&beans', 'potato', 'ragi', 'rapeseed', 'rice', 'safflower', 'sugarcane', 'sunflower', 'turmeric', 'urad', 'wheat']

pred_user=clf.predict(np.array([[20,6.7,30,17,30,60,10,9]]))
print(pred_user)

#print(np.argmax(pred_user[0]))
#print(crops[np.argmax(pred_user[0])])

greater_than_zero={}
index = -1

for i in pred_user:
    for p in i:
        index+=1
        if(p > 0):
            greater_than_zero[crops[index]] = p 
    for i in greater_than_zero:
        print("Crop:- {0}  Probab:- {1}".format(i,greater_than_zero[i]*100))
```

```python
from sklearn.metrics import accuracy_score
# Finding the accuracy of the model
a=accuracy_score(np.argmax(y_test,axis=1),np.argmax(pred,axis=1))
print("The accuracy of this model is: ", a*100)
```
